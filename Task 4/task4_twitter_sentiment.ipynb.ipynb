Twitter Sentiment Analysis using NLP

Steps:
1. Data Loading
2. Data Preprocessing
3. Feature Extraction (TF-IDF)
4. Model Training (Logistic Regression)
5. Evaluation & Insights

```python
from google.colab import files

print("Upload training file")
uploaded_train = files.upload()
print("Upload validation file")
uploaded_val = files.upload()
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
train_file = list(uploaded_train.keys())[0]
val_file = list(uploaded_val.keys())[0]

train_df = pd.read_csv(train_file, header=None)
val_df = pd.read_csv(val_file, header=None)

# Add column names
train_df.columns = ['id', 'entity', 'sentiment', 'tweet']
val_df.columns = ['id', 'entity', 'sentiment', 'tweet']

train_df.head()
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_tweet(text):
    text = str(text).lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

train_df['cleaned_text'] = train_df['tweet'].apply(clean_tweet)
val_df['cleaned_text'] = val_df['tweet'].apply(clean_tweet)
```

```python
vectorizer = TfidfVectorizer(max_features=5000)

X_train = vectorizer.fit_transform(train_df['cleaned_text'])
X_val = vectorizer.transform(val_df['cleaned_text'])

y_train = train_df['sentiment']
y_val = val_df['sentiment']
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_val)

print("Accuracy:", accuracy_score(y_val, y_pred))
print("\nClassification Report:\n", classification_report(y_val, y_pred))
```

```python
print("Sample Predictions:")
for i in range(5):
    print("Tweet:", val_df['tweet'].iloc[i])
    print("Actual:", y_val.iloc[i])
    print("Predicted:", y_pred[i])
    print("-"*50)
```

### Insights

* The model was able to classify tweets into Positive, Negative, and Neutral categories using Logistic Regression.
* Text preprocessing (removing stopwords, special characters) improved model performance.
* TF-IDF helped convert text data into meaningful numerical features.
* The model achieved a reasonable accuracy, indicating effective sentiment prediction.
* Some misclassifications may occur due to sarcasm or ambiguous language in tweets.

### Conclusion

* Logistic Regression works well for basic sentiment analysis tasks.
* Further improvements can be made using advanced models like Naive Bayes or Deep Learning.
