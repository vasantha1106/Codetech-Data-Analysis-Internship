```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Netflix Data Analysis") \
    .getOrCreate()
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = spark.read.csv("netflix_titles.csv", header=True, inferSchema=True)
df.show(5)
```

```python
df.printSchema()
df.describe().show()
```

```python
df = df.dropna()
df = df.dropDuplicates()
```

```python
df.groupBy("type").count().show()
```

```python
df.groupBy("release_year").count().orderBy("count", ascending=False).show()
```

```python
df.groupBy("rating").count().orderBy("count", ascending=False).show()
```

```python
pdf = df.toPandas()

import matplotlib.pyplot as plt

pdf['type'].value_counts().plot(kind='bar')
plt.show()
```

The analysis highlights key trends in content distribution, popular categories, and production patterns.
